```python
class CaesarCipher:
    @staticmethod
    def capturePlainText():
        a = input("Enter the message which needs to be encrypted: ")
        print("Plain text:", a)
        return a

    @staticmethod
    def encryptCaesarCipher(a, Key):
        cipher = ""
        for i in a:
            if i.isupper():
                cipherIntValue = ord(i)
                cipherIntValue = (cipherIntValue - 65 + Key) % 26 + 65
            else:
                cipherIntValue = ord(i)
                cipherIntValue = (cipherIntValue - 97 + Key) % 26 + 97
            cipherStringValue = chr(cipherIntValue)
            cipher += cipherStringValue
        print("Encrypted Cipher:", cipher)
        print("Key/Shift Value: ", Key)
        return cipher

    @staticmethod
    def decryptCaeserCipher(Et, Key):
        plainText = ""
        for i in Et:
            if i.isupper():
                cipherIntValue = ord(i)
                cipherIntValue = (cipherIntValue - 65 - Key) % 26 + 65
            else:
                cipherIntValue = ord(i)
                cipherIntValue = (cipherIntValue - 97 - Key) % 26 + 97
            cipherStringValue = chr(cipherIntValue)
            plainText += cipherStringValue
        print("Cipher:", Et)
        print("Decrypted Cipher:", plainText)
        print("Key/Shift Value: ", Key)
        return plainText

class PlayFairCipher:
    def generatePlayFairSquare(self, key):
        print("Play Fair Square")
        sqr = [[" " for _ in range(5)] for _ in range(5)]
        key = key.upper().replace("J", "I")
        key += "ABCDEFGHIKLMNOPQRSTUVWXYZ"
        key = ''.join(dict.fromkeys(key))
        
        for i in range(5):
            for j in range(5):
                sqr[i][j] = key[i * 5 + j]
        
        for row in sqr:
            print(row)       
        return sqr

    def capturePlainText(self):
        a = input("Enter the message which needs to be encrypted: ")
        print("Plain text:", a)
        return a
    
    def splitPlainText(self,pt):
        # To split the PT with a pair of 2
        pairs = [pt[i:i+2] for i in range(0, len(pt), 2)]
        print(pairs)
        return pairs

    def validatePlainText(self, pt):
        print("Plain Text:", pt)
        pairs = self.splitPlainText(pt)
        # To check if the splitted pair has the same characters
        for index, pair in enumerate(pairs):
            if len(pair) == 2 and pair[0] == pair[1]:
                # To add an 'X' as a filler variable if two pairs are the same
                updated_pair = pair[:1] + 'X' + pair[1:]
                pairs[index] = updated_pair
                break
        #     # combined_plain_text = ''.join(pairs)
        #     # print("Combined Plain Text:", combined_plain_text)
        #     # pairs = self.splitPlainText(combined_plain_text)
        print(pairs)
        
        #To resplit 1st we rejoin the string then split
        combined_plain_text = ''.join(pairs)
        print("Combined Plain Text:", combined_plain_text)
        pairs = self.splitPlainText(combined_plain_text)

        #To Add X to last character if length of the Plain Text is ODD
        length = len(combined_plain_text)
        print(length)
        if length % 2 != 0 :
            last_pair = pairs[-1]
            updated_last_pair = last_pair[:1] + 'X' + last_pair[1:]
            pairs[-1] = updated_last_pair
        print("Final Pair:",pairs)

        return pairs
    #To find the Encrypted Text from the PlayFairSquare
    def pfCEncryption(self,ptPairs,pfSqr):
        CipherList=[]
        for i in ptPairs:
            ListLoc = []
            a = i[0]
            b = i[1]
            for index1,i in enumerate(pfSqr):
                for index2,j in enumerate(i):
                    if j == a:
                        ListLoc.append([index1,index2])
                    if j == b:
                        ListLoc.append([index1,index2])
            #print(ListLoc)

            if(ListLoc[0][0] == ListLoc[1][0]):
                #print("Same Row")
                row1, col1 = ListLoc[0]
                row2, col2 = ListLoc[1]
                # To append the next character in the pfSqr of the index to the CipherList
                next_index = (ListLoc[0][1] + 1) % len(pfSqr[ListLoc[0][0]])
                CipherList.append(pfSqr[ListLoc[0][0]][next_index])
                next_index = (ListLoc[1][1] + 1) % len(pfSqr[ListLoc[0][0]])
                CipherList.append(pfSqr[ListLoc[0][0]][next_index])
            elif(ListLoc[0][1] == ListLoc[1][1]):
                #print("Same Column") 
                next_index = (ListLoc[1][0] + 1) % len(pfSqr)
                
                # Append the character in the next row but with the same column index to CipherList
                print("**",pfSqr[next_index][ListLoc[0][1]])
                CipherList.append(pfSqr[next_index][ListLoc[0][1]])
                next_index = (ListLoc[0][0] + 1) % len(pfSqr)
                CipherList.append(pfSqr[next_index][ListLoc[0][1]])
                print("***",pfSqr[next_index][ListLoc[0][1]])
            else:
                #print("Different Row / Column")
                row1, col1 = ListLoc[0]
                row2, col2 = ListLoc[1]
                col1, col2 = col2, col1
                print(pfSqr[row1][col1])
                print(pfSqr[row2][col2])
                # Append the character at the opposite corner of the rectangle to CipherList
                CipherList.append(pfSqr[row1][col1])
                CipherList.append(pfSqr[row2][col2])    
        print("CipherList:",CipherList)
        return CipherList
                        
    def pfDecryption(self,pfSqr,ptPairs):
        CipherList=[]
        for i in ptPairs:
            ListLoc = []
            a = i[0]
            b = i[1]
            for index1,i in enumerate(pfSqr):
                for index2,j in enumerate(i):
                    if j == a:
                        ListLoc.append([index1,index2])
                    if j == b:
                        ListLoc.append([index1,index2])
            #print(ListLoc)

            if(ListLoc[0][0] == ListLoc[1][0]):
                #print("Same Row")
                row1, col1 = ListLoc[0]
                row2, col2 = ListLoc[1]
                # To append the next character in the pfSqr of the index to the CipherList
                next_index = (ListLoc[0][1] - 1) % len(pfSqr[ListLoc[0][0]])
                CipherList.append(pfSqr[ListLoc[0][0]][next_index])
                next_index = (ListLoc[1][1] - 1) % len(pfSqr[ListLoc[0][0]])
                CipherList.append(pfSqr[ListLoc[0][0]][next_index])
            elif(ListLoc[0][1] == ListLoc[1][1]):
                #print("Same Column") 
                next_index = (ListLoc[1][0] - 1) % len(pfSqr)
                
                # Append the character in the next row but with the same column index to CipherLisT
                CipherList.append(pfSqr[next_index][ListLoc[0][1]])
                next_index = (ListLoc[0][0] - 1) % len(pfSqr)
                CipherList.append(pfSqr[next_index][ListLoc[0][1]])
            else:
                #print("Different Row / Column")
                row1, col1 = ListLoc[0]
                row2, col2 = ListLoc[1]

                col1, col2 = col2, col1
                print(pfSqr[row1][col1])
                print(pfSqr[row2][col2])
                # Append the character at the opposite corner of the rectangle to CipherList
                CipherList.append(pfSqr[row1][col1])
                CipherList.append(pfSqr[row2][col2])   
        print("CipherList:",CipherList)
        return CipherList

    def encryptPlainText(self, pt,Key):
        print("EPT")
        pfSqr = self.generatePlayFairSquare(Key)
        ptPairs = self.validatePlainText(pt)
        Cipher = self.pfCEncryption(ptPairs,pfSqr)
        # print(pfSqr)
        return Cipher,pfSqr

    def decryptCypherText(self,cipher,pfSqr):
        print(" decrypt ")
        print(pfSqr)
        pairs = self.splitPlainText(cipher)
        decryptedList = self.pfDecryption(pfSqr,pairs)
        if decryptedList[-1] == 'X':
            decryptedList.pop()
        print("Decrypted List:" ,decryptedList)


def displayMenu():
    print("\n********Menu**********\n1.Caeser Cipher\n2.Play Fair Cipher\n3.Exit")


def main():
    while True:
        displayMenu()
        choice = int(input("Enter your choice(1-3):"))
        if choice == 1:
            CC = CaesarCipher()
            plainText = CC.capturePlainText()
            key = int(input("Enter the Key/Shift Value(Integer Value): "))
            encryptedText = CC.encryptCaesarCipher(plainText, key)
            ptDecrypted = CC.decryptCaeserCipher(encryptedText, key)

        elif choice == 2:
            print("PlayFair Cipher")
            PFC = PlayFairCipher()
            plainText = PFC.capturePlainText()
            Key = input("Enter the Key Value:")
            encryptedText,pfSqr = PFC.encryptPlainText(plainText , Key)
            a = input("Do you want to decrypt the Encrypted Cipher? (Y/N)")
            if a == 'Y' or a == 'y':
                decryptedText = PFC.decryptCypherText(encryptedText,pfSqr)
            else:
                continue
        elif choice == 3:
            print("Exiting")
            break
        else:
            print("Invalid Choice.. Please try again :")


main()
```
